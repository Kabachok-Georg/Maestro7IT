Пример того, как можно создать файл `data.csv`, сохранить данные и затем считать их для красивого вывода с использованием Python и библиотеки `csv`.

### Шаг 1: Создание файла data.csv и запись данных

```python
import csv

# Данные для записи в CSV файл
data = [
    ["Имя", "Возраст", "Город"],
    ["Иван", 25, "Москва"],
    ["Анна", 30, "Санкт-Петербург"],
    ["Петр", 22, "Новосибирск"],
]

# Создание и запись в CSV файл
with open("data.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("Данные успешно записаны в data.csv")
```

### Шаг 2: Чтение данных из файла и красивый вывод

```python
import csv

# Чтение данных из CSV файла
with open("data.csv", mode="r", encoding="utf-8") as file:
    reader = csv.reader(file)
    data_from_file = list(reader)

# Красивый вывод
print(f"{'Имя':<10} {'Возраст':<10} {'Город':<20}")
print("-" * 40)
for row in data_from_file[1:]:  # Пропускаем заголовок
    print(f"{row[0]:<10} {row[1]:<10} {row[2]:<20}")
```

### Описание кода:

**Создание файла:**

В первом блоке создается файл data.csv, куда записываются данные.

**Чтение файла:**

Во втором блоке считываются данные из файла, и затем они красиво выводятся с выравниванием.

### Пример вывода:

```
Имя        Возраст    Город               
----------------------------------------
Иван       25         Москва             
Анна       30         Санкт-Петербург    
Петр       22         Новосибирск        
```

Этот код можно использовать для работы с CSV файлами.

````python
'''
Записываем данные в файл data.csv
'''

import csv

# Данные для записи в CSV файл
data = [
    ["Имя", "Возраст", "Город"],
    ["Иван", 25, "Москва"],
    ["Анна", 30, "Санкт-Петербург"],
    ["Петр", 22, "Новосибирск"],
    ["Георгий", 14, "Пушкин"],
    ["Максим", 19, "Краснодар"],
    ["Александр", 18, "Астрахань"],
    ["Вероника", 17, "Петропавловск-Камчатский"],
    ["Ксения", 16, "Тула"]
]

# Создание и запись в CSV файл
with open("data.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("Данные успешно записаны в data.csv")
````

````python
'''
Счмтываем данные из файла data.csv
'''

import csv

# Чтение данных из CSV файла
with open("data.csv", mode="r", encoding="utf-8") as file:
    reader = csv.reader(file)
    data_from_file = list(reader)

# Красивый вывод
print(f"{'Имя':<10} {'Возраст':<10} {'Город':<20}")
print("-" * 40)
for row in data_from_file[1:]:  # Пропускаем заголовок
    print(f"{row[0]:<10} {row[1]:<10} {row[2]:<20}")
````

### Шаг 2: Улучшенный код с построением графика

```python
import csv
import matplotlib.pyplot as plt

# Чтение данных из CSV файла
with open("data.csv", mode="r", encoding="utf-8") as file:
    reader = csv.reader(file)
    data_from_file = list(reader)

# Красивый вывод
print(f"{'Имя':<10} {'Возраст':<10} {'Город':<20}")
print("-" * 40)
names = []
ages = []
for row in data_from_file[1:]:  # Пропускаем заголовок
    print(f"{row[0]:<10} {row[1]:<10} {row[2]:<20}")
    names.append(row[0])
    ages.append(int(row[1]))

# Построение графика
plt.figure(figsize=(8, 5))
plt.bar(names, ages, color='skyblue')

# Оформление графика
plt.title("Возраст участников", fontsize=16)
plt.xlabel("Имя", fontsize=12)
plt.ylabel("Возраст", fontsize=12)

# Показываем график
plt.show()
```

### Описание кода:

**Чтение данных:**

Данные из файла data.csv читаются и выводятся в консоль.
Имена и возраста заносятся в отдельные списки для построения графика.

**Построение графика:**

Используется библиотека `matplotlib` для построения столбчатой диаграммы, где имена участников отображаются по `оси X`, а возраст по `оси Y`.

**Оформление графика:**

Добавляются заголовки, подписи к осям для более красивого и понятного отображения.

**Пример вывода**

**Таблица в консоли:**
```
Имя        Возраст    Город               
----------------------------------------
Иван       25         Москва             
Анна       30         Санкт-Петербург    
Петр       22         Новосибирск        
```

**График:**

Вы увидите столбчатую диаграмму с именами и возрастом участников.

### Результат:

Этот код читает данные, выводит их в консоль, а затем строит наглядный график, отображающий возраст участников.

````python
'''
Работа с библиотекой matplotlib.
Построение графика столбчатой диаграммы.
'''

import csv
import matplotlib.pyplot as plt

# Чтение данных из CSV файла
with open("data.csv", mode="r", encoding="utf-8") as file:
    reader = csv.reader(file)
    data_from_file = list(reader)

# Красивый вывод
print(f"{'Имя':<10} {'Возраст':<10} {'Город':<20}")
print("-" * 40)
names = []
ages = []
for row in data_from_file[1:]:  # Пропускаем заголовок
    print(f"{row[0]:<10} {row[1]:<10} {row[2]:<20}")
    names.append(row[0])
    ages.append(int(row[1]))

# Построение графика
plt.figure(figsize=(8, 5))

# Устанавливаем ширину столбцов и отступы между ними
plt.bar(names, ages, color='skyblue', width=0.6)

# Оформление графика
plt.title("Возраст участников", fontsize=16)
plt.xlabel("Имя", fontsize=12)
plt.ylabel("Возраст", fontsize=12)

# Добавление отступов между столбцами за счёт настройки меток оси X
plt.xticks(rotation=0, ha='center', fontsize=10)

# Оптимизация размещения элементов графика
plt.tight_layout()

# Добавление отступов снизу (нижний край графика)
plt.subplots_adjust(bottom=0.2)  # Увеличение нижнего отступа

# Показываем график
plt.show()
````
